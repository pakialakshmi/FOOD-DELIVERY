```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
locations = ["Chennai", "Bangalore", "Hyderabad", "Coimbatore", "Kochi"]
items = ["Pizza", "Burger", "Biryani", "Dosa", "Fried Rice", "Pasta", "Shawarma"]

data = []

for i in range(1, 121):
    hour = random.randint(8, 22)
    minute = random.randint(0, 59)

    item = random.choice(items)

    if item == "Pizza":
        amount = 250
    elif item == "Burger":
        amount = 180
    elif item == "Biryani":
        amount = 300
    elif item == "Dosa":
        amount = 120
    elif item == "Fried Rice":
        amount = 220
    elif item == "Pasta":
        amount = 280
    else:
        amount = 160

    data.append([
        i,
        f"{hour:02d}:{minute:02d}",
        random.choice(locations),
        item,
        amount
    ])
```

```python
df = pd.DataFrame(
    data,
    columns=["Order_ID", "Order_Time", "Location", "Item", "Amount"]
)

print("DATASET PREVIEW")
print(df.head())
```
